### load dict

```python
dt_pt = open('all.dict', 'r')
vocab = json.load(dt_pt)
dt_pt.close()
```

```python
print(len(vocab))
print(vocab['英语'])
```

```python
def id2tk(vb):
    dict_ = {}
    for i in vb.items():
        dict_[i[1]]=i[0]
    return dict_

id_to_tk = id2tk(vocab)
len(id_to_tk)
```

```python
lda = joblib.load('lda.pkl')
```

```python
topic_token = lda.components_
```

```python
main_id = exp0.argsort()[-50:]
```

```python
main_tk = []
for i in main_id:
    main_tk.append(id_to_tk[i])
print(main_tk)
```

```python
def print_tp_main_tks(id_):
    exp = topic_token[id_]
    main_id = exp.argsort()[-50:]
    main_tk = []
    for i in main_id:
        main_tk.append(id_to_tk[i])
    print(main_tk)
```
